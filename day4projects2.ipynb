```python
import pandas as pd

data = {
    "Student":["Ali","Sara","Ahmed","Ayesha","Usman","Hina","Bilal","Zara",
               "Hamza","Noor","Asad","Maham","Daniyal","Iqra","Farhan"],

    "Department":["CS","CS","SE","SE","IT",
                  "IT","CS","SE","IT","CS",
                  "SE","IT","CS","SE","IT"],

    "Gender":["Male","Female","Male","Female","Male",
              "Female","Male","Female","Male","Female",
              "Male","Female","Male","Female","Male"],

    "Semester":[4,5,6,4,5,
                6,7,8,4,5,
                6,7,8,4,5],

    "CGPA":[3.4,3.8,3.2,3.9,3.1,
            3.6,3.7,3.5,3.0,3.9,
            3.3,3.8,3.6,3.7,3.4]
}

df = pd.DataFrame(data)
print(df)
```

```python
Department_wise_std_counts = df.groupby("Department")["Student"].count()
print(Department_wise_std_counts)
```

```python
Department_wise_avg_cgpa = df.groupby("Department")["CGPA"].mean()
print(Department_wise_avg_cgpa)
```

```python
best_performing_department = Department_wise_avg_cgpa.idxmax()
print("Best Performing Department:", best_performing_department)
```

```python
male_female_counts = df.groupby("Gender")["Student"].count()
print(male_female_counts)
```

```python
department_Gender_counts = df.groupby(["Department","Gender"])["Student"].count()
print(department_Gender_counts)
```

```python
semester_wise_avg_cgpa = df.groupby("Semester")["CGPA"].mean()
print(semester_wise_avg_cgpa)
```

```python
report = df.groupby("Department").agg(
    Average_CGPA=("CGPA","mean"),
    Max_CGPA=("CGPA","max"),
    Min_CGPA=("CGPA","min"),
    Student_Count=("Student","count")
)

print(report)
```
